```python
# Creating Class with name ITEM for creating Inventory
class Item:
    def __init__(self, name, price, stock, item_id):
        
        # Name of the Product
        self.name = name
        
        #Price of the product
        self.price = price
        
        #Initial Inventory of the product
        self.stock = stock
        
         #Item Id of the Product
        self.item_id = item_id
        
    # Function for updating the stock
    def updateStock(self, stock):
        self.stock = stock

    # Function for buying the stock
    def buyFromStock(self):
        if self.stock == 0:
            # raise not item exception
            pass
        self.stock -= 1

        
    
# Creating class with name VendingMachine
class VendingMachine:
    def __init__(self):
        self.amount = 0
        self.items = []
        
    # Adding items created using Item Class objects
    def addItem(self, item):
        self.items.append(item)

    # Function for displaying the Items & their details
    def showItems(self):
        print('\nCurrent Inventory is:\n***************')

        for item in self.items:      
            if item.stock == 0:
                self.items.remove(item)
        for item in self.items:
            print("Item name: ",item.name, ", Item ID: ",item.item_id, ", Price:", item.price, "cents, Current stock: ",item.stock)

        print('***************\n')
        
    # For adding cash
    def addCash(self, money):
        self.amount = self.amount + money
        print("Cash of ",money," cents added")
        print("Current Cash is: ", self.amount, " cents")
        
    # Function for Buying the item
    def buyItem(self, item):
        if self.amount < item.price:
            print('You can\'t but this item. Insert more coins.')
        else:
            self.amount -= item.price
            item.buyFromStock()
            print('You got ' +item.name)
            print('Cash remaining: ' + str(self.amount))
            
    # Function to check if the User Entered Item exist in stock or not
    def containsItem(self, wanted):
        ret = False
        for item in self.items:
            if item.item_id == wanted:
                ret = True
                break
        return ret

    # Function to get Item detail of user entered Item
    def getItem(self, wanted):
        ret = None
        for item in self.items:
            if item.item_id == wanted:
                ret = item
                break
        return ret

    # Function for adding additional cash if the price of item bought is more than the current cash
    def insertAmountForItem(self, item):
        price = item.price
        while self.amount < price:
            print("Alert!! You current cash is less. Please add " + str(price - self.amount) + " cents more. ")
            self.amount = self.amount + float(input('insert ' + str(price - self.amount) + ' cents more : '))

    # Function for checking any refund left & processing the refund
    def checkRefund(self):
        if self.amount > 0:
            print(self.amount, "cents refunded.")
            self.amount = 0

        print('Thank you, have a nice day!\n')
```

```python
print("------- Vending Machine with Python-------\n")
print("----------The Items In Stock Are----------\n")

machine = VendingMachine()

# Creating Items list, their price & quantity at start
item1 = Item('Slim Bar', 5,10,"A")
item2 = Item('Far Bar',10,10,"B")
item3 = Item('Mar Bar',20,10,"C")
item4 = Item('Twin Bar',50,10,"D")

# adding items details to Machine Class objects
machine.addItem(item1)
machine.addItem(item2)
machine.addItem(item3)
machine.addItem(item4)

money =0 

no_of_coins = int(input('Enter how many coins you want to Insert: '))
value_of_coins = int(input('Enter the value of coins (only in multiples of 5,10, 20 or 50 only - it will be same for all the coins): '))

right_denomination = False
while (right_denomination==False):
    if ((value_of_coins==5 or value_of_coins==10 or value_of_coins==20 or value_of_coins==50)):
        right_denomination = True
        break
    else:
        print("Sorry, Machine accepts coins with denomination of 5,10, 20 or 50 only")
        value_of_coins = int(input(("Enter the value of coins (only in multiples of 5,10, 20 or 50 only) again")))
                         
money = no_of_coins * value_of_coins

machine.addCash(money)

continueToBuy = True
while continueToBuy == True:
    machine.showItems()
    selected = input('Enter the item ID you want to buy: ')
    if machine.containsItem(selected):
        item = machine.getItem(selected)
        machine.insertAmountForItem(item)
        machine.buyItem(item)

        a = input('buy something else? (y/n): ')
        if a == 'n':
            continueToBuy = False
            machine.checkRefund()
        else:
            continue

    else:
        print('Item not available. Select another item.')
        continue
```
